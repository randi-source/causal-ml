### 3. Symmetry
In linear equation lets say, we might say that Symptom severity equals to Disease severity and other factors. When it is using algebra, the equation can be changed into Disease severity equals to Symptom severity minus other factors, and this statement is indeed does not make sense.



### Causality
- X causes Y if when all confounders are adjusted
- an intervention in X results in a change in Y
- but intervention in Y does not change X



### How to represent causality?
1. Directed Acyclic Graphs (DAGs) \
Which is a part of graph theory that depict causality in graph. Here, we have vertices (bullet) and edges (the arrow). In this schema, we can depict vertices as the variables and edges as the causal direction (Directed), which in practice, cannot pointed back into the source vertices (Acyclic).
2. Structural Equation Model (SEMs) \
Which similar to linear equation but has slightly different sign (":="), which mean it cannot be inverted like W:=f1(X) cannot be transformed into X:=f1(W)



## Causal Inference
Answering questions based on causal connections



### 3 Gifts of Causal Inference
1. Do-Operator
2. Deconfounding Confounding
3. Estimating Causal Effects

### 1. Do-Operator
It is simulate physical intervention on Do factor
- Sometimes, we can't intervene with the real world case, because it is physically impossible or unethical to do like for example giving a pack of ciggaretes to people to measure the response in disease of lung desease is unethical
- In real world, we dont have the way to collect data about interventional probability distribution, that is when we dont have data about probability that include the do operator
- In this regards, the Rules od Do-Calculus may provide a way to re-express nor rewrite the probability that we were interested in but cannot be measure directly



### Gift 2: (Deconfounding) Confounding
Anything that makes P(Y|do(X)) different than P(Y|X)
- Anything that makes the probability distribution of P(Y|do(X)) given intervention of X is different from proability distribution given observation of X in P(Y|X)
- for example, if we knew that age is confounder, this can help inform our analysis
- Let's say if we want to assess the the impact of education on income, if we don't take age into confounder, the naive things to do, we can separate the analysis based on each education level and just compare the difference in income



- But, when we consider age as confounders, we should not compare data between age group, but rather, compare data within age group
